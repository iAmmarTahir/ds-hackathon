```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://www.washington.edu/students/timeschd/SPR2024/'
```

```python
response = requests.get(url)
department_links = []
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    

    # Find all <ul> elements
    ul_elements = soup.find_all('ul')
    
    # Iterate through each <ul> element
    for ul in ul_elements:

        # Find all <li> elements within the current <ul>
        li_elements = ul.find_all('li')
        
        # Iterate through each <li> element
        for li in li_elements:
            # Find all <a> elements within the current <li>
            a_elements = li.find_all('a')
            
            # Iterate through each <a> element and print its href attribute
            for a in a_elements:
                # Check if the <a> tag has an 'href' attribute
                if 'href' in a.attrs and '.html' in a['href']:
                    department_links.append(a['href'])
                    
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
# Filtering the unwanted links
department_links = department_links[6:]
```

```python
tabular_data = []
for link in department_links:
    print("Fetching: ", link)
    full_url = 'https://www.washington.edu/students/timeschd/SPR2024/' + link
    res = requests.get(full_url)
    if res.status_code == 200:
        soup = BeautifulSoup(res.content, 'html.parser')
        # Find all the tables on the page
        tables = soup.find_all('table', attrs={'bgcolor': None})
        for table in tables:
            table_data = []
            rows = table.find_all('tr')
            for row in rows:
                cells = row.find_all('td')
                for cell in cells:
                    pre_tags = cell.find_all('pre')
                    if len(pre_tags) > 0:
                        pre_text = pre_tags[0].get_text(strip=True)
                        tabular_data.append(pre_text)
```

```python
import re
```

```python
columns = ['SLN', 'Days', 'Time']
restructured_data = []
for elem in tabular_data: 
    elem_list = [item for item in elem.split(' ') if item != '']
    filtered = [elem_list[0], elem_list[2], re.split(r'[A-Za-z]', elem_list[3], maxsplit=1)[0]]
    restructured_data.append(filtered)
    
df = pd.DataFrame(restructured_data, columns=columns)
    
df
```

```python
days = [item for item in df['Days'].unique() if item.isalpha()]
days = [item for item in days if item[0].isupper()]
days = [item for item in days if item != 'VAR']
days
```

```python
df = df[df['Days'].isin(days)]
```

```python
def expand_days(row):
    days = row['Days']
    expanded_days = []
    if 'M' in days:
        expanded_days.append('Monday')
    if 'T' in days:
        expanded_days.append('Tuesday')
    if 'W' in days:
        expanded_days.append('Wednesday')
    if 'Th' in days:
        expanded_days.append('Thursday')
    if 'F' in days:
        expanded_days.append('Friday')
    return pd.Series(expanded_days, dtype=str)
```

```python
expanded_df = df.apply(expand_days, axis=1).stack().reset_index(level=1, drop=True).to_frame('Day')
result_df = df.drop('Days', axis=1).join(expanded_df)
result_df = result_df.reset_index()
```

```python
result_df = result_df.drop('index', axis=1)
```

```python
result_df[['start_time', 'end_time']] = result_df['Time'].str.split('-', expand=True)
result_df = result_df.drop('Time', axis = 1)
```

```python
result_df['start_time'] = result_df['start_time'].str.zfill(4)
result_df['end_time'] = result_df['end_time'].str.zfill(4)
```

```python
result_df['start_time'] = pd.to_datetime(result_df['start_time'], format='%H%M').dt.time
result_df['end_time'] = pd.to_datetime(result_df['end_time'], format='%H%M').dt.time
```

```python
# Function to add 'AM' or 'PM' based on the hour range
def add_am_pm(time):
    hour = int(time.hour)
    if hour in [8, 9, 10, 11]:
        return str(time) + ' AM'
    else:
        return str(time) + ' PM'
result_df['start_time'] = result_df['start_time'].apply(add_am_pm)
result_df['end_time'] = result_df['end_time'].apply(add_am_pm)
```

```python
from datetime import datetime, time
# Convert start_time and end_time columns to Time datatype
result_df['start_time'] = result_df['start_time'].apply(lambda x: datetime.strptime(x, '%I:%M:%S %p').time())
result_df['end_time'] = result_df['end_time'].apply(lambda x: datetime.strptime(x, '%I:%M:%S %p').time())
```

```python
result_df.head(10)
```

```python
# Create time bins of 1 hour for start_time and end_time
result_df['start_time_bins'] = pd.cut(pd.to_datetime(result_df['start_time'], format='%H:%M:%S').dt.hour, bins=range(0, 25, 1), right=False)
result_df['end_time_bins'] = pd.cut(pd.to_datetime(result_df['start_time'], format='%H:%M:%S').dt.hour, bins=range(0, 25, 1), right=False)

# Group data by Day and start_time_bins, and calculate the count of unique SLN values
start_time_counts = result_df.groupby(['Day', 'start_time_bins'])['SLN'].nunique()

# Group data by Day and end_time_bins, and calculate the count of unique SLN values
end_time_counts = result_df.groupby(['Day', 'end_time_bins'])['SLN'].nunique()

# Create the chart for each day
days = result_df['Day'].unique()
```

```python
days
```

```python
for day in ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']:
    # Filter data for the current day
    start_time_day_counts = start_time_counts.loc[day]
    end_time_day_counts = end_time_counts.loc[day]
    
    # Create the chart
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))
    
    # Plot start_time counts
    axes[0].bar(start_time_day_counts.index.astype(str), start_time_day_counts.values)
    axes[0].set_xlabel('Start Time')
    axes[0].set_ylabel('Count')
    axes[0].set_title('Start Time Counts')
    
    # Rotate x-labels for start_time
    for tick in axes[0].get_xticklabels():
        tick.set_rotation(45)
    
    # Plot end_time counts
    axes[1].bar(end_time_day_counts.index.astype(str), end_time_day_counts.values)
    axes[1].set_xlabel('End Time')
    axes[1].set_ylabel('Count')
    axes[1].set_title('End Time Counts')
    
    # Rotate x-labels for start_time
    for tick in axes[1].get_xticklabels():
        tick.set_rotation(45)
    
    # Set tight layout
    plt.tight_layout()
    fig.suptitle(day)
    
    # Show the chart
    plt.show()
```
